# ScaFi on jupyter notebook
A PoF for ScaFi script running on a notebook. It should be useful for learning purpose

## ScaFi dependecy

```scala
import $ivy.`it.unibo.scafi::scafi-core:0.3.3`
import $ivy.`it.unibo.scafi::scafi-simulator:0.3.3`
```

## Incarnation definition

```scala
import it.unibo.scafi.config.GridSettings
import it.unibo.scafi.incarnations.BasicAbstractSpatialSimulationIncarnation
import it.unibo.scafi.lib.StandardLibrary
import it.unibo.scafi.space.Point3D
import java.time.Instant

object MyIncarnation extends BasicAbstractSpatialSimulationIncarnation with StandardLibrary {
  override implicit val idBounded: MyIncarnation.Builtins.Bounded[Int] = Builtins.Bounded.of_i
  override type P = Point3D
}
```

## Simulator definition

```scala
import MyIncarnation._
val range = 90
val step = 60
val simulation = MyIncarnation.simulatorFactory.gridLike(GridSettings(stepx = step, stepy = step), rng = range)
```

## Program definition

```scala
val program = new AggregateProgram {
  override def main(): Any = rep(0)(_ + 1)
}
```

## Program execution

```scala
simulation.exec(program)
```
